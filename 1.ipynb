```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
thetas = np.linspace(0,1,1001)
```

```python
def likelihood(r):
    if r == 0:
        return 1 - thetas # P(r=0|theta)
    elif r == 1:
        return thetas # P(r=1|theta)
```

```python
def get_posterior_distribution(r,prior_distribution):
    lp = likelihood(r) * prior_distribution # P(r|theta) * P(theta)
    return lp / lp.sum()
```

```python
intial_theta = np.array([1/len(thetas) for _ in thetas])
```

```python
intial_theta
```

```python
p_theta = get_posterior_distribution(1,intial_theta)
```

```python
p_theta
```

```python
plt.plot(thetas,p_theta)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.show()
```

```python
clicks,noclicks = 2,38
p_theta = intial_theta
```

```python

for _ in range(clicks):
    p_theta = get_posterior_distribution(1,p_theta)

for _ in range(noclicks):
    p_theta = get_posterior_distribution(0,p_theta)
```

```python
def show_figure(thetas,brief_rate_distributions):
    for brief_rate_distribution in brief_rate_distributions:
        plt.plot(thetas,brief_rate_distribution)
    plt.xlabel(r'$\theta$')
    plt.ylabel(r'$p(\theta)$')
    plt.show()
```

```python
show_figure(thetas,[p_theta])
```
